```python
import pandas as pd
import requests
```

```python
#Get data By API
response = requests.get("https://dummyjson.com/products")
```

```python
#all response in json file
response.json()
```

```python
#convert into data frame
df = pd.DataFrame(response.json()["products"])
df.head(3)
```

```python
# if we want specific column only then:
response = requests.get("https://dummyjson.com/products/category/smartphones")
df = pd.DataFrame(response.json()["products"])[["title", "price", "brand"]]
df
```

```python
#if we have a lot of pages and we want to collect API response then we need to it in a loop:
for i in range(1,10):
  response = requests.get(f"https://dummyjson.com/products?limit=10&skip={i*10}")
  temp_df = pd.DataFrame(response.json()["products"])
  df = df.append(temp_df, ignore_index = True)
df
```

```python
#export data

df.to_csv("smartphones.csv")
```
