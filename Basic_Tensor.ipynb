```python
import numpy as np
```

```python
# 0D/scalar:

scalar = np.array(5)
print("scalar: ",scalar)
print("rank/array dimension: ", scalar.ndim)
print("shape", scalar.shape)
```

```python
vector = np.array([2,4,5,6])
print("array: ",vector)
print("rank/No of axis/array dimension: ", vector.ndim)
print("shape", vector.shape)

# Note:
# 1D Tensor = Vector --> [2,3,4,5] and dimension of this vector is 4.(vector elements a.k.a dimension).
# array dimension and vector dimension both are different things.
```

```python
# 2D/Matrix:

matrix = np.array([[2,3,4],[2,1,4],[1,6,4]])
print("matrix",matrix)
print("rank/array dimension: ", matrix.ndim)
print("shape", matrix.shape)
```

```python
# 3D:

# A 3D list with dimensions 2x2x3 (2 layers, 2 rows, 3 columns)
_3d_list = np.array([
    [[1, 2, 3], [4, 5, 6]],     # Layer 0
    [[7, 8, 9], [10, 11, 12]]   # Layer 1
])


print(_3d_list)
print("rank/array dimension: ", _3d_list.ndim)
print("shape", _3d_list.shape)
```
